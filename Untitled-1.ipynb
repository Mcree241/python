```python
# Titanic Data Analysis Example

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
# Titanic dataset available online from seaborn
df = sns.load_dataset("titanic")

# 3. Explore Data
print("Shape of dataset:", df.shape)
print("\nInfo:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
print("\nFirst 5 rows:")
print(df.head())

# 4. Basic Data Analysis
print("\nSurvival counts:")
print(df['survived'].value_counts())

print("\nClass distribution:")
print(df['class'].value_counts())

print("\nCorrelation matrix (numeric only):")
print(df.corr(numeric_only=True))

# 5. Visualizations

# Survival distribution
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df, palette="viridis")
plt.title("Survival Distribution")
plt.show()

# Class vs Survival
plt.figure(figsize=(6,4))
sns.countplot(x="class", hue="survived", data=df, palette="mako")
plt.title("Passenger Class vs Survival")
plt.show()

# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df['age'].dropna(), bins=20, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 6. Findings/Observations
print("\nFINDINGS & OBSERVATIONS:")
print("- Most passengers did not survive (majority survival=0).")
print("- Higher survival rate observed in First Class compared to Third Class.")
print("- Younger passengers (children) had a better chance of survival.")
print("- Fare is positively correlated with survival (wealthier passengers had higher chances).")
```
